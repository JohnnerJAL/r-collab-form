```r
# Convierte a datos numéricos (si están como texto)
datos$Tiempo <- as.numeric(datos$Tiempo)
datos$Estimados <- as.numeric(datos$Estimados)
datos$Participados <- as.numeric(datos$Participados)
datos$Invitaciones <- as.numeric(datos$Invitaciones)
datos$Docentes <- as.numeric(datos$Docentes)
```

**Tabla de frecuencia de ultimas columnas**

*   Tiempo: '6. ¿Cuánto tiempo podrás dedicar semanalmente a un semillero de
investigación?'
*   Estimados: '7. ¿Cuantos semilleros de investigación crees que tiene la Universidad Pedagógica y Tecnológica de Colombia?'
*   Participados: '8. En cuantos semilleros has participado?'
*   Invitaciones: '9.  Â¿Cuantas veces has recibido invitaciones o información directa acerca de los semilleros de investigacion?'

*   Docentes: '10. ¿Cuántos docentes te han hablado directamente sobre la
posibilidad de participar en un semillero de investigación?'

```r
# tabla de frecuencia (variable cuantitativa ordinaria)
# Tiempo
# '6. ¿Cuánto tiempo podrás dedicar semanalmente a un semillero de investigación?'

Freq(datos$Tiempo, main = "Tiempo dedicado (hrs/sem)")

# Tiempo dedicado semanalmente
freq_tiempo <- Freq(datos$Tiempo, plot = FALSE)
porcentajes <- as.numeric(gsub("%", "", freq_tiempo$perc))  # Convertir "50%" a 50
barplot(porcentajes,
        names.arg = freq_tiempo$x,
        main = "Tiempo dedicado semanalmente (%)",
        ylab = "Porcentaje", xlab = "Horas",
        col = "skyblue")
```

La mayoría de los estudiantes (58.5%) están dispuestos a dedicar entre 0 y 2 horas semanales a un semillero. Solo un 2.4% consideraría dedicar más de 10 horas.

```r
# Función para graficar frecuencias en porcentaje
graficar_freq <- function(variable, titulo, color) {
  freq <- Freq(variable, plot = FALSE)
  porcentajes <- as.numeric(gsub("%", "", freq$perc))  # quitar símbolo %
  barplot(porcentajes,
          names.arg = freq$x,
          main = titulo,
          ylab = "Porcentaje",
          xlab = "Valor",
          col = color)
}
```

```r
# tabla de frecuencia (variable cuantitativa ordinaria)
# Estimados
# '7. ¿Cuantos semilleros de investigación crees que tiene la Universidad Pedagógica y Tecnológica de Colombia?'

Freq(datos$Estimados, main = "Número estimado de semilleros")
graficar_freq(datos$Estimados, "Número estimado de semilleros (%)", "salmon")
```

El 70.7% de los estudiantes estima que la universidad tiene entre 0 y 20 semilleros. Solo un pequeño grupo cree que existen más de 100.

```r
# tabla de frecuencia (variable cuantitativa ordinaria)
# Participados
# '8. En cuantos semilleros has participado?'

Freq(datos$Participados, main = "Número de semilleros en los que participó")
graficar_freq(datos$Participados, "Semilleros en los que participó (%)", "lightgreen")
```

El 97.6% de los estudiantes no ha participado en ningún semillero. Solo uno reporta alguna participación.

```r
# tabla de frecuencia (variable cuantitativa ordinaria)
# Invitaciones
# '9.  Â¿Cuantas veces has recibido invitaciones o información directa acerca de los semilleros de investigacion?'

Freq(datos$Invitaciones, main = "Veces que recibió invitaciones")
graficar_freq(datos$Invitaciones, "Número de invitaciones recibidas (%)", "lightblue")
```

Casi un tercio de los estudiantes ha recibido solo una invitación sobre semilleros, y más de la mitad ha recibido tres o menos.

```r
# tabla de frecuencia (variable cuantitativa ordinaria)
# Docentes
# '10. ¿Cuántos docentes te han hablado directamente sobre la posibilidad de participar en un semillero de investigación?'

Freq(datos$Docentes, main = "Número de docentes que lo invitaron")
graficar_freq(datos$Docentes, "Docentes que lo invitaron (%)", "orange")
```

El 34.1% de los estudiantes ha sido contactado por solo un docente, mientras que menos del 10% ha recibido invitación directa de más de 7 docentes.
